```python
'''   A, B and C are classes 
A is a super class. B is a sub class of A. C is a sub class of B.  

Create three methods in each class, 2 methods are specific to each class and third 
method (override method) should be in all three Classes A, B and C 

Create a class with main method. Create an object for each class A, B and C in main 
method and call every method of each class using its own object/instance. 

Call an overridden method with super class reference to B and C classâ€™s objects 
Runtime Polymorphism with Data Members/Instance variables, Repeat the above 
process only for data members  '''
```

```python
class A:
    def __init__(self):
        self.a_value='Value from A '
    def method_a1(self):
        return "Method A1 in class A"
    def method_a2(self):
        return "Method A2 in class A"
    def overriden_method(self):
        return "Overriden method in class A"
    
class B(A):
    def __init__(self):
        super().__init__()
        self.b_value='Value from B '
    def method_b1(self):
        return "Method B1 in class B"
    def method_b2(self):
        return "Method B2 in class B"
    def overriden_method(self):
        return "Overriden method in class B"
    
class C(B):
    def __init__(self):
        super().__init__()
        self.c_value='Value from C '
    def method_c1(self):
        return "Method C1 in class C"
    def method_c2(self):
        return "Method C2 in class C"
    def overriden_method(self):
        return "Overriden method in class C"
    
class Main:
    a=A()
    b=B()
    c=C()
     
    #calling methods of class A
    print("\n Class A Methods")
    print(a.method_a1())
    print(a.method_a2())
    print(a.overriden_method())
    
    #calling methods of class B
    print("\n Class B Methods")
    print(b.method_b1())
    print(b.method_b2())
    print(b.overriden_method())
    
    #calling methods of class C
    print("\n Class C Methods")
    print(c.method_c1())
    print(c.method_c2())
    print(c.overriden_method())
    
    # Runtime Polymorphism
    print("\nCalling overridden methods using superclass reference:")
    super_b=A()
    super_b.overriden_method=b.overriden_method
    print(super_b.overriden_method())
    
    print("\n")
    super_c=A()
    super_c.overriden_method=c.overriden_method
    print(super_c.overriden_method())
    
    print("\n Accessing Data members:")
    print(a.a_value)
    print(b.b_value)
    print(c.c_value)
    
    print("\n Accessing Data members using super class keyword.")
    
    super_b_value=b.a_value
    print("Accessing A's value from B object:",super_b_value)
    
    super_c_value=c.b_value
    print("Accessing B's value from C object:",super_c_value)
    
if __name__=="__main__()":
    main_instance=Main()
    main_instance.main()
```
